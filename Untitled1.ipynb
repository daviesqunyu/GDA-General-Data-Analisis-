```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set clean chart style globally
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.spines.top']   = False
plt.rcParams['axes.spines.right'] = False

# County employee dataset
data = {
    'name':       ['Alice','Brian','Carol','David','Esther',
                   'Felix','Grace','Hassan','Irene','James',
                   'Kezia','Leonard','Mary','Noel','Olive'],
    'department': ['Health','Health','Health','Education','Education',
                   'Finance','Finance','ICT','Agriculture','Agriculture',
                   'Public Works','Public Works','ICT','Health','Education'],
    'salary':     [85000,72000,68000,91000,77000,
                   95000,88000,105000,62000,58000,
                   79000,83000,97000,55000,71000],
    'ward':       ['Cherangany','Saboti','Kiminini','Kwanza','Cherangany',
                   'Saboti','Kiminini','Kwanza','Cherangany','Saboti',
                   'Kiminini','Kwanza','Cherangany','Saboti','Kwanza'],
    'years_exp':  [6,5,7,4,6,8,3,9,3,4,7,6,5,2,4],
    'active':     [1,1,1,1,1,1,0,1,1,1,1,1,1,0,1]
}

df = pd.DataFrame(data)
print("Dataset ready!", df.shape)
df.head()
```

```python
dept_counts = df['department'].value_counts()

plt.figure(figsize=(10, 5))
plt.bar(dept_counts.index, dept_counts.values, color='steelblue', edgecolor='white')

plt.title('Staff Count per Department', fontsize=14, fontweight='bold')
plt.xlabel('Department')
plt.ylabel('Number of Staff')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

```python
avg_sal = df.groupby('department')['salary'].mean().sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.bar(avg_sal.index, avg_sal.values,
              color=['#2196F3','#4CAF50','#FF9800','#E91E63','#9C27B0','#00BCD4'],
              edgecolor='white', width=0.6)

# Add value labels on top of each bar
for bar in bars:
    ax.text(bar.get_x() + bar.get_width()/2,
            bar.get_height() + 1000,
            f'KES {bar.get_height():,.0f}',
            ha='center', va='bottom', fontsize=9)

ax.set_title('Average Salary by Department — Trans Nzoia County',
             fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Department')
ax.set_ylabel('Average Salary (KES)')
ax.set_ylim(0, avg_sal.max() * 1.2)
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

```python
avg_sal = df.groupby('department')['salary'].mean().sort_values()

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.barh(avg_sal.index, avg_sal.values,
               color='#1976D2', edgecolor='white', height=0.6)

# Value labels on the right of each bar
for bar in bars:
    ax.text(bar.get_width() + 500, bar.get_y() + bar.get_height()/2,
            f'KES {bar.get_width():,.0f}',
            va='center', fontsize=9)

ax.set_title('Average Salary by Department', fontsize=14, fontweight='bold')
ax.set_xlabel('Average Salary (KES)')
ax.set_xlim(0, avg_sal.max() * 1.25)
plt.tight_layout()
plt.show()
```

```python
dept_counts = df['department'].value_counts()

colors = ['#2196F3','#4CAF50','#FF9800','#E91E63','#9C27B0','#00BCD4']

fig, ax = plt.subplots(figsize=(8, 6))
wedges, texts, autotexts = ax.pie(
    dept_counts.values,
    labels=dept_counts.index,
    autopct='%1.1f%%',           # show percentage on each slice
    colors=colors,
    startangle=90,
    pctdistance=0.85,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)

# Make percentage text bold
for autotext in autotexts:
    autotext.set_fontweight('bold')

ax.set_title('Staff Distribution by Department', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.hist(df['salary'], bins=8, color='#1976D2',
        edgecolor='white', alpha=0.85)

# Add mean and median lines
ax.axvline(df['salary'].mean(),   color='red',    linestyle='--', linewidth=1.5, label=f'Mean: KES {df["salary"].mean():,.0f}')
ax.axvline(df['salary'].median(), color='orange', linestyle='--', linewidth=1.5, label=f'Median: KES {df["salary"].median():,.0f}')

ax.set_title('Salary Distribution — County Staff', fontsize=14, fontweight='bold')
ax.set_xlabel('Salary (KES)')
ax.set_ylabel('Number of Employees')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

# Colour points by department
departments = df['department'].unique()
colors = ['#2196F3','#4CAF50','#FF9800','#E91E63','#9C27B0','#00BCD4']

for i, dept in enumerate(departments):
    subset = df[df['department'] == dept]
    ax.scatter(subset['years_exp'], subset['salary'],
               label=dept, color=colors[i], s=100, edgecolors='white', linewidth=0.8)

# Add trend line
z = np.polyfit(df['years_exp'], df['salary'], 1)
p = np.poly1d(z)
x_line = np.linspace(df['years_exp'].min(), df['years_exp'].max(), 100)
ax.plot(x_line, p(x_line), color='gray', linestyle='--', linewidth=1.2, label='Trend')

ax.set_title('Years of Experience vs Salary', fontsize=14, fontweight='bold')
ax.set_xlabel('Years of Experience')
ax.set_ylabel('Salary (KES)')
ax.legend(bbox_to_anchor=(1.01, 1), loc='upper left', fontsize=9)
plt.tight_layout()
plt.show()
```

```python
# Select numeric columns only
numeric_df = df[['salary', 'years_exp', 'active']]
corr = numeric_df.corr()

fig, ax = plt.subplots(figsize=(6, 5))
sns.heatmap(corr,
            annot=True,          # show numbers inside cells
            fmt='.2f',            # 2 decimal places
            cmap='Blues',         # colour scheme
            vmin=-1, vmax=1,       # correlation always -1 to 1
            linewidths=0.5,
            ax=ax)

ax.set_title('Correlation Heatmap — Numeric Variables',
             fontsize=13, fontweight='bold', pad=12)
plt.tight_layout()
plt.show()
```
